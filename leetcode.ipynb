```python
class Solution(object):
    def longestCommonPrefix(self, sts):
        l=""
        n=min(sts, key=len)
    # length=0
        i=0
        while i!=len(n):
            curr=n[i]
            
            t=0
            for s in sts:
                if s[i]==curr:
                    t+=1
            if t==len(sts):
                i+=1
                l+=curr
            else :
                break
        return l
                    
            
            
            
```

```python
class Solution(object):
    def smallestEvenMultiple(self, n):
        if n%2==0:
            return n
        else:
            return n*2
```

```python
class Solution(object):
    def isPalindrome(self, x):
            if 0>x:
                     return False
            return int(str(x)[::-1])==x
```
